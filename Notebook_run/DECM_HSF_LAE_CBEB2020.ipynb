<font size =3><div style="text-align: justify"> This interactive tool makes it possible to investigate how the motor neuron (MN) firing properties is influenced by changes in monoamnergic drive and, biophysical and morphological MN characteristics. The two-compartment MN (represented by one soma and one dendrite) implemented at this work followed the Hodgkin-Huxley (1952) formalism. Also, the dynamics of the conductances were based on Traub (1977) and Dai et al. (2002). The interactive tool has three different alpha MNs: S (slow fatigue-resistant), FR (fast fatigue-resistant) and FF (fast fatigable). All the equations utilized can be found inside the napp.mod and caL.mod mechanisms. </div>


## First of all, import the library and functions necessary to run this interactive tool.

```python
from ipywidgets import interact_manual
import ipywidgets as widgets
import widget_params, params2
```

<font size=3
&nbsp;</p>
To observe the MN responses, it is possible to inject a current pulse into the soma compartment. To configure the pulse, set the following widgets: 
<p>&nbsp;</p>

* Current delay [ms] (simulation delay before stimulus is applied)


* Depolarizing current pulse duration [ms]


* Time Between Injected Current [ms] (time between depolarizing and hyperpolarizing pulses)


* Hyperpolarizing current pulse duration [ms]


* Depolarizing current pulse amplitude [nA]


* Hyperpolarizing current pulse amplitude [nA]
<p>&nbsp;</p> 

  
The dendrite MN morphology and biophysic properties can also be altered by the widgets:
<p>&nbsp;</p>
* Dendrite lenght [mm]


* Dendrite Diameter [um] 


* Persistent sodium conductance [mS/cm²] (part of the persistent inward current conductance (PIC), is located in the soma of the MN)


* Calcium type-L conductance [mS/cm²] (part of the persistent inward current conductance (PIC), is located in the dendrite of the MN)


* Slow potassium conductance [mS/cm²] (the condutance responsible for after-hyperpolarization dynamics)


* Monoaminergic drive (to regulate the L-type calcium conductance. It comprehends values between 0.0 and 1.0. The 0.0 value indicates a passive dendrite, 0.2 represents anesthetised animal data and, 1.0 refers to decerebrated animal data, which implies in monoaminergic function near to its maximal activity). 


# Pulse current injection and MNs parameters configuration

<font size=3> <div style="text-align: justify">Set the widget values and click "Run Interact". If the initial configuration is kept, the FF MN under hyperexcitable ALS stage will be represented based on Matoso, Fernandes, Elias (2018). Note the FF MN self-firing even without the depolarizing current pulse injection (between 2000 ms and 4000 ms). This FF MN stops firing solely after a hyperpolarizing current pulse is injected (near 5000 ms). To observe the resulting mean MN firing frequency and number of action potentials open the file "pulse_injection.csv".</div>

```python
pulse=params2.params_init()
parameters = widget_params.widget_params_interact_manual_pulse()
_ = widgets.interact_manual(pulse.view_pulse,mutype= parameters[0],gama = parameters[1], delay =parameters[2],
              predur = parameters[3],
              meddur = parameters[4],posdur = parameters[5],preamp = parameters[6],
              posamp = parameters[7],ld = parameters[8],diam=parameters[9], gnap=parameters[10], gcal=parameters[11], gks= parameters[12])
```

<font size =3> <div style="text-align: justify"><p>&nbsp;</p>
To observe the MN dynamic behaviour it is possible to inject a slow triangular current into the soma compartment. After that, the frequency-current relationship (f-I) can gives us some information about MN behaviour. Also, the specific widgets permits the configuration of the triangular current injection: </div>

<p>&nbsp;</p>

* Ascending triangular current duration [ms]


* Descending triangular current duration [ms]


* Ascending triangular current amplitude [nA]


* Descending triangular current amplitude [nA]

<p>&nbsp;</p>
  
<font size = 3><div style="text-align: justify">The dendrite morphology and biophysic MN properties can be altered by the widgets at the same way mentioned for the pulse current configuration.</div>

# Triangular current injection and MNs parameters configuration

<p>&nbsp;</p>
<font size = 3><div style="text-align: justify">Set the widget values and click "Run Interact". If the initial configuration is unchanged, the hysteresis behaviour appears for the FF MN firing, representing the hyperexcitable stage of the progression of ALS based on Matoso, Fernandes, Elias (2018). The frequency and the current deltas at healthy condition were paremeterized based on Lee and Heckman (1998) and Elias and Kohn (2013). To observe the resulting frequency and the current deltas and the number of action potentials open the file "triangular_injection.csv"</div>

```python
triangular=params2.params_init()
parameters = widget_params.widget_interact_manual_triangular()
_ = widgets.interact_manual(triangular.view_triangular,mutype= parameters[0],gama = parameters[1],
                            delay =parameters[2],predur = parameters[3],posdur = parameters[4],
                            posamp = parameters[5],possamp = parameters[6],ld = parameters[7],
                            diam=parameters[8], gnap=parameters[9], gcal=parameters[10], gks= parameters[11])
```
